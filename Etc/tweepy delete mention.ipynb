```python
import tweepy
```

```python
C_KEY = "-"
C_SECRET = "-"
A_TOKEN = "-"
A_SECRET = "-"
```

```python
AUTH = tweepy.OAuthHandler(C_KEY, C_SECRET)
AUTH.set_access_token(A_TOKEN, A_SECRET)

API = tweepy.API(AUTH, wait_on_rate_limit=True,
                 wait_on_rate_limit_notify=True)

ME = API.me()
MY_NAME = '@' + (ME.screen_name).lower()

print(" * name: " + MY_NAME + " *")
```
