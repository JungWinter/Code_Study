```python
X = d['data']
Y = d['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
X2_train, X2_test, Y_train, Y_test = train_test_split(X2, Y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
slr.fit(X2_train, Y_train)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X2_test, Y_test,  color='black')
plt.plot(X2_test, slr.predict(X2_test), color='blue', linewidth=3)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
for i in range(10):
    X2 = X[:, i].reshape(-1, 1)
    X2_train, X2_test, Y_train, Y_test = train_test_split(X2, Y, test_size=0.2)

    slr = LinearRegression()
    slr.fit(X2_train, Y_train)
    plt.scatter(X2_test, Y_test)
    plt.plot(X2_test, slr.predict(X2_test), linewidth=1)
plt.xticks(())
plt.yticks(())
plt.show()
```
